## Calculus / Statistical Functions

```python
a1 = 2*np.random.randn(10000) + 10
```

```python
np.mean(a1)
```

```python
np.std(a1)
```

```python
np.percentile(a1, 80)
```

```python
x = np.linspace(1,10,100)
y = 1/x**2 * np.sin(x)
dydx = np.gradient(y, x)
y_int = np.cumsum(y) * (x[1]-x[0])
```

```python
plt.plot(x, y)
plt.plot(x, dydx)
plt.plot(x, y_int)
```

```python
N=10000
x = np.linspace(1,10,N+1)
y = np.exp(-x/10)*np.sin(x)
plt.plot(x,y)
```

```python
np.mean(y[(x>=4)*(x<=7)])
np.std(y[(x>=4)*(x<=7)])
```

```python
np.percentile(y[(x>=4)*(x<=7)], 80)
```

```python
plt.plot(x, np.gradient(y, x))
```

```python
dydx = np.gradient(y,x)
```

```python
x[1:][dydx[1:]*dydx[:-1]<0]
```
