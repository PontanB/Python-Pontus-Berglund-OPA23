### a) Change name on the diffrent subjects

```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
def changename(df):
    x = 2
    y = 0
    names = ["Totalt (A-F)","Flickor (A-F)","Pojkar (A-F)","Totalt (A-E)","Flickor (A-E)","Pojkar (A-E)","Totalt (poäng)","Flickor (poäng)","Pojkar (poäng)"]
    while x < 11:
        df = df.rename({df.columns[x]: names[y]}, axis=1)
        y += 1
        x += 1
    return df
  
```

```python


df_english = pd.read_excel("../data/riket2023_åk9_np.xlsx", skiprows=8)
df_english = df_english.rename({"Riket": "Plats",},axis=1)
df_english
```

```python

df_english = changename(df_english)
df_english

```

### b) How many F in math 

```python


def count_F_first_row(df):
    
    total_A_F = pd.to_numeric(df.iloc[0]['Totalt (A-F)'])
    flickor_A_F = pd.to_numeric(df.iloc[0]['Flickor (A-F)'])
    pojkar_A_F = pd.to_numeric(df.iloc[0]['Pojkar (A-F)'])
    total_A_E = pd.to_numeric(df.iloc[0]['Totalt (A-E)'])
    flickor_A_E = pd.to_numeric(df.iloc[0]['Flickor (A-E)'])
    pojkar_A_E = pd.to_numeric(df.iloc[0]['Pojkar (A-E)'])
    
   
    total_F = round(total_A_F - (total_A_F * (total_A_E / 100)))
    flickor_F = round(flickor_A_F - (flickor_A_F * (flickor_A_E / 100)))
    pojkar_F = round(pojkar_A_F - (pojkar_A_F * (pojkar_A_E / 100)))
    
    return total_F, flickor_F, pojkar_F

total_F, flickor_F, pojkar_F = count_F_first_row(df_math)


print("Total number of F:s :", total_F)
print("Total number of F:s for girls :", flickor_F)
print("Total number of F:s for boys:", pojkar_F)
```
