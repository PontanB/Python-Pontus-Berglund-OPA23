### a) Change name on the diffrent subjects

```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
def changename(df):
    x = 2
    y = 0
    names = ["Totalt (A-F)","Flickor (A-F)","Pojkar (A-F)","Totalt (A-E)","Flickor (A-E)","Pojkar (A-E)","Totalt (poäng)","Flickor (poäng)","Pojkar (poäng)"]
    while x < 11:
        df = df.rename({df.columns[x]: names[y]}, axis=1)
        y += 1
        x += 1
    return df
```

```python


df_english = pd.read_excel("../data/riket2023_åk9_np.xlsx", skiprows=8)
df_english = df_english.rename({"Riket": "Plats",},axis=1)
df_english
```

```python

df_english = changename(df_english)
df_english

```

```python
import numpy as np
def Count_F(df):
    x = 3
    y = 6
    f = np.zeros_like(df.iloc[:, 0])  # Skapar en array med nollor med samma form som den första kolumnen i df
    while x < 6:
        column_x = df.iloc[:, x].astype(float).to_numpy()  # Konverterar kolumnen till en NumPy-array med flyttal
        column_y = df.iloc[:, y].astype(float).to_numpy()  # Konverterar kolumnen till en NumPy-array med flyttal
        f += column_x - column_y
        y += 1
        x += 1
    return f
```

```python
result = Count_F(df_math)
print(result)
```
