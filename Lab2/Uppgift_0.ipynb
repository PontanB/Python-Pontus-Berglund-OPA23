### a) Change name on the diffrent subjects

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def changename(df):
    x = 2
    y = 0
    names = ["Totalt (A-F)","Flickor (A-F)","Pojkar (A-F)","Totalt (A-E)","Flickor (A-E)","Pojkar (A-E)","Totalt (poäng)","Flickor (poäng)","Pojkar (poäng)"]
    while x < 11:
        df = df.rename({df.columns[x]: names[y]}, axis=1)
        y += 1
        x += 1
    return df
  
```

```python


df_english = pd.read_excel("../data/riket2023_åk9_np.xlsx", skiprows=8)
df_english = df_english.rename({"Riket": "Plats",},axis=1)
df_english
```

```python

df_english = changename(df_english)
df_english

```

```python
df_math = pd.read_excel("../data/riket2023_åk9_np.xlsx", skiprows=8, sheet_name= "Matematik")
df_math = df_math.rename({"Riket": "Plats",},axis=1)
df_math
```

```python
df_math = changename(df_math)
df_math
```

```python
df_swedish = pd.read_excel("../data/riket2023_åk9_np.xlsx", skiprows=8, sheet_name= "Svenska")
df_swedish = df_swedish.rename({"Riket": "Plats",},axis=1)
df_swedish
```

```python
df_swedish = changename(df_swedish)
df_swedish
```

```python
df_second_languge = pd.read_excel("../data/riket2023_åk9_np.xlsx", skiprows=8, sheet_name= "Svenska som andraspråk")
df_second_languge = df_second_languge.rename({"Riket": "Plats",},axis=1)
df_second_languge
```

```python
df_second_languge = changename(df_second_languge)
df_second_languge
```

### b) How many F in math 

```python


def count_F_first_row(df):
    
    total_A_F = pd.to_numeric(df.iloc[0]['Totalt (A-F)'])
    flickor_A_F = pd.to_numeric(df.iloc[0]['Flickor (A-F)'])
    pojkar_A_F = pd.to_numeric(df.iloc[0]['Pojkar (A-F)'])
    total_A_E = pd.to_numeric(df.iloc[0]['Totalt (A-E)'])
    flickor_A_E = pd.to_numeric(df.iloc[0]['Flickor (A-E)'])
    pojkar_A_E = pd.to_numeric(df.iloc[0]['Pojkar (A-E)'])
    
   
    total_F = round(total_A_F - (total_A_F * (total_A_E / 100)))
    flickor_F = round(flickor_A_F - (flickor_A_F * (flickor_A_E / 100)))
    pojkar_F = round(pojkar_A_F - (pojkar_A_F * (pojkar_A_E / 100)))
    
    return total_F, flickor_F, pojkar_F

total_F, flickor_F, pojkar_F = count_F_first_row(df_math)


print("Total number of F:s :", total_F)
print("Total number of F:s for girls :", flickor_F)
print("Total number of F:s for boys:", pojkar_F)
```

### c) diagrams of total points

```python
df_Poäng = pd.concat([df_english["Totalt (poäng)"].iloc[[0]], 
                           df_math['Totalt (poäng)'].iloc[[0]], 
                           df_swedish['Totalt (poäng)'].iloc[[0]], 
                           df_second_languge['Totalt (poäng)'].iloc[[0]]], 
                           axis=1, keys=['Engelska', 'Matte', 'Svenska', 'SSAS'])

print(df_Poäng)
```

```python

subjects = ["Engelska", "Matte", "Svenska", "SSAS"]
positions = range(len(subjects))

# Hjälp från Youtube
for i, subject in enumerate(subjects):
    plt.bar(positions[i], df_Poäng[subject][0])  
plt.xticks(positions, subjects)
plt.xlabel("Ämnen")
plt.ylabel("Poäng")
plt.title("Poäng för olika ämnen")

plt.show()
```

### d) subplots of huvudmän

```python
df_huvudman_Poäng = pd.concat([df_english["Totalt (poäng)"], 
                           df_math['Totalt (poäng)'], 
                           df_swedish['Totalt (poäng)'], 
                           df_second_languge['Totalt (poäng)']], 
                           axis=1, keys=['Engelska', 'Matte', 'Svenska', 'SSAS'])
df_huvudman_Poäng
```

```python
index_replacements = {0: "samtliga", 1: "Kommunal", 2: "Enskild", 3: "Skolverket"}

# Använd rename() för att byta ut indexnamnen
df_Poäng_renamed = df_huvudman_Poäng.rename(index=index_replacements)

# Visa den nya DataFrame med de ändrade indexnamnen
print(df_Poäng_renamed)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
# Help from chatgpt
df_huvudman_Poäng["SSAS"] = pd.to_numeric(df_huvudman_Poäng["SSAS"], errors="coerce").fillna(0)
colors = ["yellow", "teal", "violet", "orange"]

df_huvudman_Poäng["Engelska"].plot(kind="bar", ax=axes[0, 0], title="Engelska", color=colors)
df_huvudman_Poäng["Matte"].plot(kind="bar", ax=axes[0, 1], title="Matte", color=colors)
df_huvudman_Poäng["Svenska"].plot(kind="bar", ax=axes[1, 0], title="Svenska", color=colors)
df_huvudman_Poäng["SSAS"].plot(kind="bar", ax=axes[1, 1], title="SSAS", color=colors)

for ax in axes.flat:
    ax.set_xticklabels(df_Poäng_renamed.index, rotation=0, ha="center")
plt.tight_layout()
plt.show()
```
